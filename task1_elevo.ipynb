## import lib

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
import seaborn as sns
```

## load and prepare the data

```python
#load data
df = pd.read_csv("https://raw.githubusercontent.com/sarthak-srivastava/Student-Performance-Dataset/refs/heads/master/student-por.csv")

#feauture selection
df = df[[ 'studytime', 'G3']]

#split data
x=df.drop('G3',axis=1)
y= df['G3']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=10)
```

## data visiualization

```python
print(df.shape)
df.info()
df.describe()
plt.scatter(df.studytime,df.G3,color='b')
plt.xlabel('study time')
plt.ylabel('final exam scoreG3')
```

## applaing model

```python
linr = LinearRegression()
linr.fit(x_train,y_train)
```

```python
y_predict = linr.predict(x_test)
```

```python
plt.scatter(x_test,y_test,label = 'actuall')
plt.scatter(x_test,y_predict,label='prediction')
plt.title('Actual vs Predicted Scores')
plt.xlabel('Study Time')
plt.ylabel('Final Score')
plt.legend()
plt.grid()
```

## evaluate the model

```python
mae = mean_absolute_error(y_test,y_predict)
mse = mean_squared_error(y_test,y_predict)

print('mean_absolute_error: ',  mae)
print("mean_squared_error: ", mse)
```
